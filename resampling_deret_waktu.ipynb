```python
# Resampling pada time series data

import pandas as pd
import numpy as np
```

```python
n_rows = 364*24
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns = cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df
```

```python
df.resample('M')['col1'].sum().to_frame()
```
