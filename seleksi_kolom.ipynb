```python
# Seleksi kolom & baris menggunakan loc
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)

df
```

```python
df.loc[[0,3,4], ['B','E']]
```

```python
df.loc[df['B']>10, ['B','D','E']]
```

```python
df.loc[0:4, 'B':'D']
```
